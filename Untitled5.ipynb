```python
def load_docs():
    with open("C:\\Users\\shiva\\OneDrive\\Desktop\\project\\doc1.md", "r", encoding="utf-8") as f:
        doc1 = f.read()

    with open("C:\\Users\\shiva\\OneDrive\\Desktop\\project\\doc2.md", "r", encoding="utf-8") as f:
        doc2 = f.read()

    with open("C:\\Users\\shiva\\OneDrive\\Desktop\\project\\doc3.md", "r", encoding="utf-8") as f:
        doc3 = f.read()

    return [doc1, doc2, doc3]


docs = load_docs()

print("Number of documents:", len(docs))
print("\nSample content:\n")
print(docs[0][:500])
```

```python
def chunk_text(text, chunk_size=80):
    words = text.split()
    chunks = []

    for i in range(0, len(words), chunk_size):
        chunk = " ".join(words[i:i+chunk_size])
        chunks.append(chunk)

    return chunks


# Apply chunking again
all_chunks = []

for doc in docs:
    chunks = chunk_text(doc)
    all_chunks.extend(chunks)

print("Total chunks created:", len(all_chunks))
print("\nSample chunk:\n")
print(all_chunks[0])
```

```python
from sentence_transformers import SentenceTransformer

# Load embedding model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
embeddings = embedding_model.encode(all_chunks)

print("Embedding shape:", embeddings.shape)
```

```python
import faiss
import numpy as np

# Convert embeddings to numpy array
embedding_array = np.array(embeddings)

# Get dimension
dimension = embedding_array.shape[1]

# Create FAISS index
index = faiss.IndexFlatL2(dimension)

# Add embeddings to index
index.add(embedding_array)

print("Total vectors in FAISS index:", index.ntotal)
```

```python
def retrieve(query, k=5):
    query_embedding = embedding_model.encode([query])
    distances, indices = index.search(np.array(query_embedding), k)

    results = []
    scores = []

    for dist, i in zip(distances[0], indices[0]):
        chunk = all_chunks[i]
        if len(chunk.strip()) > 50:
            results.append(chunk)
            scores.append(round(float(dist), 4))

    return results, scores


# Test retrieval
query = "What is escrow payment?"

results = retrieve(query)

print("🔍 Retrieved Chunks:\n")
for i, chunk in enumerate(results):
    print(f"Chunk {i+1}:\n{chunk}\n")
```

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

# -------------------------------
# LOAD ENV (SECURITY)
# -------------------------------
load_dotenv()

api_key = os.getenv("OPENROUTER_API_KEY")

if not api_key:
    raise ValueError("OPENROUTER_API_KEY not found. Please set it in .env file")

client = OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1"
)

# -------------------------------
# GENERATION FUNCTION
# -------------------------------
def generate_answer(query, context_chunks):

    context = "\n\n".join(context_chunks)

    prompt = f"""
You are an AI assistant for a construction company.

STRICT RULES:
- Answer ONLY from the context below
- Do NOT use external knowledge
- Do NOT guess
- If answer is not present, say "Not available in context"

Context:
{context}

Question:
{query}

Answer:
"""

    response = client.chat.completions.create(
        model="meta-llama/llama-3-8b-instruct",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    return response.choices[0].message.content

# -------------------------------
# TEST QUERY
# -------------------------------
query = "What is escrow payment?"

# IMPORTANT: handle your updated retrieve() return
retrieved_chunks, scores = retrieve(query, k=5)  # Assuming retrieve() now returns both chunks and scores

print("📄 Retrieved Context:\n")
for i, (c, s) in enumerate(zip(retrieved_chunks, scores)):
    print(f"Chunk {i+1} (score: {s}):\n{c[:200]}\n")

answer = generate_answer(query, retrieved_chunks)

print("\n💡 FINAL ANSWER:\n")
print(answer)
```

```python
import ollama

def generate_answer_ollama(query, context_chunks):
    
    context = "\n\n".join(context_chunks)
    
    prompt = f"""
You are a strict AI assistant.

RULES:
1. Answer ONLY using the provided context
2. Do NOT add any external knowledge
3. Do NOT explain beyond the context
4. If answer is not clearly present, say: "I don't know"

Context:
{context}

Question:
{query}

Answer:
"""
    
    response = ollama.chat(
        model='phi',
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    
    return response['message']['content']
```

```python
query = "What is escrow payment?"

# ✅ unpack properly
chunks, scores = retrieve(query)

print("📄 Retrieved Context:\n")

for i, (c, s) in enumerate(zip(chunks, scores)):
    print(f"Chunk {i+1} (score: {s}):\n{c[:200]}\n")

# ✅ pass only chunks to model
answer = generate_answer_ollama(query, chunks)

print("\n💡 Ollama Answer:\n")
print(answer)
```
